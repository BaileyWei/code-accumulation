```python
sample_l = []
label2id = {
    'argument-subsection': 1,
    'argument-section': 2
}
def generate_out_text(data, label2id, typ=0):
    if typ == 0:
        input_text = data['data']['text']
        
        try:
            entity_list = data['annotations'][0]['result']
        except:
            entity_list = data['predictions'][0]['result'] 
        mapping = []
        for e in entity_list:
            if e['type'] == 'labels':
                if e['value']['labels'][0] in list(label2id.keys()):
                    mapping.append((e['value']['start'], e['value']['end'], e['value']['labels'][0]))
        mapping.sort(key=lambda x: x[:2])
        for i,tuple_ in enumerate(mapping):
            input_before = input_text[:tuple_[0]+i*7]
            add = f"<{label2id[tuple_[2]]}>{input_text[tuple_[0]+i*7:tuple_[1]+i*7]}</{label2id[tuple_[2]]}>"
            input_after = input_text[tuple_[1]+i*7:]
            input_text = input_before+add+input_after
        return input_text
    if typ == 1:
        input_text = data['data']['text']
        entity_list = data['annotations'][0]['result']
        label = [0 for i in range(len(input_text.split()))]
        mapping = []
        for e in entity_list:
            if e['type'] == 'labels':
                if e['value']['labels'][0]in list(label2id.keys()):
                    s_span = len(input_text[:e['value']['start']].split()) 
                    e_span = s_span + len(input_text[e['value']['start']:e['value']['end']].split())
                    mapping.append((s_span, e_span, e['value']['labels'][0]))
        mapping.sort(key=lambda x: x[:2])
#         print(mapping)
        label = [0 for i in range(mapping[0][0])]
        for i,tuple_ in enumerate(mapping):
            if i != 0:
                label.extend([0 for i in range(tuple_[0]-mapping[i-1][1])])
            label.extend([label2id[tuple_[2]] for i in range(tuple_[1]-tuple_[0])])
        label.extend([0 for i in range(len(input_text.split())-mapping[-1][1])])
        return ', '.join([str(l) for l in label])
    
    if typ == 2:
        input_text = data['data']['text']
        entity_list = data['annotations'][0]['result']
        mapping = defaultdict(list)
        for e in entity_list:
            if e['type'] == 'labels':
                if e['value']['labels'][0] in list(label2id.keys()):
                    mapping[e['value']['labels'][0]].append((e['value']['start'], e['value']['end']))
        
        out = []
        for k,v in mapping.items():
            l = []
            for t in v:
                l.append(f'string {input_text[t[0]:t[1]]} index {str(t[0])};{str(t[1])}')
            out.append(f'{k} entity: ' + ', '.join(l))
        return '\n'.join(out)
            
def add_space_before_punctuation(text):
    text1 = re.sub(r'(?<=[^\s\\>)])\((?=[^\s])', r' (', text)
    text2 = re.sub(r'(?<=[^\s])\,', r' ,', text1)
    text3 = re.sub(r'\§(?=([0-9A-Za-z]+))', r'§ ', text2)
    text4 = re.sub(r':-', r' :-', text3)
    if text4.endswith(':') and not text4.endswith(' :'):
        text4 = text4[:-1] + ' :'
    if text4.endswith('.') and not text4.endswith(' .'):
        text4 = text4[:-1] + ' .'
    text5 = re.sub(r'(?<=[^\s])\;', r' ;', text4)
    return text5
                
        
def get_sorted_entity_dict(data, label2id, label=True, typ=0):
    entity_dict = dict()
    if label==True:
        try:
            entity_list = data['annotations'][0]['result']
        except:
            entity_list = data['predictions'][0]['result'] 
        input_text = data['data']['text']
        mapping = []
        for e in entity_list:
            if e['type'] == 'labels':
                if e['value']['labels'][0] in list(label2id.keys()):
                    mapping.append((e['value']['start'], e['value']['end'], e['value']['labels'][0]))
        mapping.sort(key=lambda x: x[:2])
        for i,tuple_ in enumerate(mapping):
            s = tuple_[0]
            e = tuple_[1]
            entity_dict['{};{}'.format(str(s), str(e))] = tuple_[2]
    else:
        entity_tuple_l = []
        text = add_space_before_punctuation(data)
        for k,v in label2id.items():
            pattern = "(?<=<{}>).+?(?=</{}>)".format(str(v), str(v))
            entity_tuple_l.extend([(m.start(0), m.end(0), k) for m in re.finditer(pattern,text)])
        entity_tuple_l.sort(key=lambda x: x[:2])
        for i,tuple_ in enumerate(entity_tuple_l):
            s = tuple_[0]-3-i*7
            e = tuple_[1]-3-i*7
            entity_dict['{};{}'.format(str(s), str(e))] = tuple_[2]
    return entity_dict

def compute_metrics(label, predict):
    hit = 0
    for k,v in label.items():
        if predict.get(k) == v:
            hit += 1
    label_num = len(label)
    predict_num = len(predict)
    return hit, label_num, predict_num
        
    
for data in dataset:
    text = data['data']['text']
    entity_list = data['annotations'][0]['result']
    output = generate_out_text(data, label2id, typ=2)
    string = "Input: {}\nOutput: {}\n".format(text, output)
    sample_l.append(string)
```

```python
# [
#     {
#         "text": "Subject to article 3, the day appointed for the coming into force of— section 15(1) (Repeals) and the entries relating to:— section 78 of the Companies Act 1989 contained in Schedule 5, is 2nd April 2001.",
#         "jx": "NY"
#         "relation_list": [
#             {
#                 "subject": "2nd April 2001",
#                 "object": "section 15(1)",
#                 "subj_char_span": [
#                     189,
#                     203
#                 ],
#                 "obj_char_span": [
#                     70,
#                     83
#                 ],
#                 "predicate": ["appl-full"],
#                                 "sub_type": "para",
#                 "obj_type": "para"
#             }
#         ],
#         "entity_list": [
#             {
#                 "text": "article 3",
#                 "type": "para-ref",
#                 "char_span": [
#                     11,
#                     20
#                 ]
#             },
#             {
#                 "text": "section 15(1)",
#                 "type": "para",
#                 "char_span": [
#                     70,
#                     83
#                 ]
#             },
#             {
#                 "text": "2nd April 2001",
#                 "type": "date-exp",
#                 "char_span": [
#                     189,
#                     203
#                 ]
#             }
#         ]
#     },
# ]
def label_studio_2_norm(data, labeled=True):
    data_dict = {}
    data_dict['text'] = data['data']['text']
    data_dict['jx'] = data['data']['jx']
    data_dict['id'] = data['id']
    relation_list = []
    entity_list = []
    mapping = {}
    if labeled == True:
        result = data['annotations'][0]['result']
    else:
        data_dict['relation_list'] = relation_list
        data_dict['entity_dict'] = entity_list
        return data_dict
    
    for d in result:
        # entity
        if d.get('value'):
            text = d['value']['text']
            typ = d['value']['labels'][0]
            s = d['value']['start']
            e = d['value']['end']
            if d['value']['text'].startswith(' '):
                s = s + 1
            if d['value']['text'].endswith(' '):
                e = e - 1
            entity_dict = {
                "text": data['data']['text'][s:e],
                "type": typ,
                "char_span": [s, e]
            }
            mapping[d['id']] = [[s, e], data['data']['text'][s:e], typ]
            entity_list.append(entity_dict)
    for d in result:
        if d.get('from_id'):
            relation_dict = {
                "subject": mapping[d['from_id']][1],
                "object": mapping[d['to_id']][1],
                "subj_char_span": mapping[d['from_id']][0],
                "obj_char_span": mapping[d['to_id']][0],
                "predicate": d['labels'],
                "sub_type": mapping[d['from_id']][2],
                "obj_type": mapping[d['to_id']][2]

            }
            relation_list.append(relation_dict)
    data_dict['relation_list'] = relation_list
    data_dict['entity_dict'] = entity_list
    return data_dict

if __name__ == "__main__":
    res_list = []
    issue = []
    for i,data in enumerate(labeled_data_m):
        try:
            data_dict = label_studio_2_norm(data)
        
            res_list.append(data_dict)
        except:
            issue.append(i)
```

```python
def get_template_embeds(template_dataset, temp_emb_path):
#     temp_emb_path = f'/data/workspace/weiyubai/dataset/code_update/labeled/labeled_emb_{encoder_name}.pkl'
    if not os.path.exists(temp_emb_path):
        template_embeddings = list()
        for record in tqdm(template_dataset):
            template_embeddings.append(get_sentence_embeds(record))
        template_embeddings_mat = torch.concat(template_embeddings)
        pickle.dump(template_embeddings_mat, open(temp_emb_path, 'wb'))
    else:
        template_embeddings_mat = pickle.load(open(temp_emb_path, 'rb'))

    return template_embeddings_mat

def get_sentence_embeds(text):
    encoded_input = tokenizer(text, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        model_output = model(**encoded_input)
    # Perform pooling
    sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

    # Normalize embeddings
    sentence_embeddings = F.normalize(sentence_embeddings, p=2, dim=1)
    return sentence_embeddings

def mean_pooling(model_output, attention_mask):
    sequence_output = model_output.last_hidden_state
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(sequence_output.size()).float()
    return torch.sum(sequence_output * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

def find_top_k(test_text, support_texts, encoder, temp_emb_path, k=3, mask_duplicated=False):
    support_embeds = get_template_embeds(support_texts, temp_emb_path)
    test_embed = get_sentence_embeds(test_text)
    rank = torch.matmul(support_embeds, test_embed.squeeze())
    rank_idx = torch.sort(rank)[1][-k:]
    new_rank_idx = []
    if mask_duplicated:
        new_rank_idx = []
        for idx in rank_idx:
            if support_texts[idx] == test_text:
                continue
            new_rank_idx.append(idx)
        if len(new_rank_idx) < k:
            new_rank_idx.extend(torch.sort(rank)[1][-(2*k-len(new_rank_idx)):-k])
        return new_rank_idx
                
    return rank_idx

def get_support_samples_by_jxs(support_dataset):
    support_by_jxs = defaultdict(list)
    for data in support_dataset:
        support_by_jxs[data['jx']].append(data)
    return support_by_jxs
    
# if __name__ == "__main__":
#     import time
#     import random
# #     support_by_jxs = get_support_samples_by_jxs(support_dataset)
#     idx = random.sample(range(len(support_dataset)), 1)
#     test_text = support_dataset[idx[0]]['text']
#     test_jx = support_dataset[idx[0]]['jx']
#     support_texts = [data['text'] for data in support_dataset]
#     s = time.time()
#     rank_idx = find_top_k(test_text, support_texts, encoder, mask_duplicated=True)
#     e = time.time()
#     print((e-s)*1000)
#     print(test_text)
#     print('-'*100)
#     for r in rank_idx:
#         sample = support_dataset[r]
#         print(sample['text'])
#     print('*'*100)
```

```python
@retry(tries=3, delay=2)
def gpt3test(prompt):
    response = openai.Completion.create(
        engine="in-context-learning",
        prompt=prompt,
        temperature=0,
        max_tokens=450,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
        stop=["\n\n"]
    )
    # result = response["choices"][0]["text"].strip()
    # result = set(result.split(";"))
    result = response["choices"][0]["text"].strip()
    return result
```

```python
def build_prompt(start_prompt, sample, typ=1):
    text = sample["text"]
    entities, relations = sample['entity_dict'], sample['relation_list']
    
    start_prompt += f"input:\n{text}\n"
    start_prompt += f"output:\n"
    if typ == 1:
        for entity in entities:
            entity_text = entity['text']
            entity_type = label2symbol[entity['type']]
#             entity_type = entity['type'].split('-')[-1]
            start_prompt += f"{entity_text}<E>{entity_type};"
        for relation in relations:
            subject = relation['subject']
            object_ = relation['object']
            predicate = label2symbol[relation['predicate']]
            start_prompt += f"{subject}<S>{object_}<O>{predicate}<P>;"
            
    if typ == 2:
        entity_dict = defaultdict(list)
        relation_dict = defaultdict(list)
        
        for entity in entities:
            entity_dict[entity['type']].append(entity['text'])
        for relation in relations:
            "<subject>{}; <object>{}".format(relation['subject'], relation['object'])
            relation_dict[relation['predicate']].append("<subject>{} and <object>{}".format(relation['subject'], relation['object']))
            
        for ent_key in entity_label.keys():
            t = '; '.join(entity_dict[ent_key])
            start_prompt += f"{ent_key} entity: {t}\n"
        for re_key in relation_label.keys():
            t = '; '.join(relation_dict[re_key])
            start_prompt += f"{re_key} relation: {t}\n"
            
        
    start_prompt += '\n\n###\n\n'
    
    return start_prompt
```

```python
# 处理GPT3生成后的结果，用于计算metrics
def postprocess(pred_results):
    results = []
    for result in pred_results:
        cur_result = defaultdict(list)
        result_seps = result.strip(';').split(';')
        for sep in result_seps:
            if '<E>' in sep:  # 实体
                cur_result['entity_list'].append(tuple(sep.split('<E>')))
            else:  # 关系
                if '<S>' in sep and '<O>' in sep and '<P>' in sep:
                    sub = sep.split('<S>')[0].strip()
                    rest = sep.split('<S>')[1]
                    obj = rest.split('<O>')[0]
                    rest = rest.split('<O>')[1]
                    rel = rest.replace('<P>', '').strip('.')
                    cur_result['relation_list'].append((sub, rel, obj))
                else:
                    cur_result['entity_list'], cur_result['relation_list'] = [], []
        results.append(cur_result)
    return results
```

```python
def calculate_metrics(pred_results, truth_results):
    metrics = defaultdict(int)
    # truth_results = remove_bad_case(truth_results)    assert len(pred_results) == len(truth_results)
    for pred, truth in zip(pred_results, truth_results):
        ent_truth = set()
        entity_truth = truth['entity_list']
        for ent in entity_truth:
            ent_truth.add((ent['text'], ent['type']))
        metrics['ent_tp'] += len(set(pred['entity_list']) & ent_truth)
        metrics['ent_fp'] += len(set(pred['entity_list']) - ent_truth)
        metrics['ent_fn'] += len(ent_truth - set(pred['entity_list']))
        rel_truth = set()
        relation_truth = truth['relation_list']
        for rel in relation_truth:
            rel_truth.add((rel['subject'], rel['predicate'][1], rel['object']))
        metrics['rel_tp'] += len(set(pred['relation_list']) & rel_truth)
        metrics['rel_fp'] += len(set(pred['relation_list']) - rel_truth)
        metrics['rel_fn'] += len(rel_truth - set(pred['relation_list']))
    ent_precision = metrics['ent_tp'] / (metrics['ent_tp'] + metrics['ent_fp'])
    ent_recall = metrics['ent_tp'] / (metrics['ent_tp'] + metrics['ent_fn'])
    ent_f1 = 2 * ent_precision * ent_recall / (ent_precision + ent_recall)
    rel_precision = metrics['rel_tp'] / (metrics['rel_tp'] + metrics['rel_fp'])
    rel_recall = metrics['rel_tp'] / (metrics['rel_tp'] + metrics['rel_fn'])
    rel_f1 = 2 * rel_precision * rel_recall / (rel_precision + rel_recall)
    metrics['ent_precision'], metrics['ent_recall'], metrics['ent_f1'] = ent_precision, ent_recall, ent_f1    
    metrics['rel_precision'], metrics['rel_recall'], metrics['rel_f1'] = rel_precision, rel_recall, rel_f1    
    return metrics

def analysis_bad_case(pred_results, truth_results):
    bad_cases = defaultdict(dict)
    # truth_results = remove_bad_case(truth_results)    assert len(pred_results) == len(truth_results)
    for pred, truth in zip(pred_results, truth_results):
        text = truth['text']
        entity_truth = truth['entity_list']
        ent_truth = set()
        for ent in entity_truth:
            ent_truth.add((ent['text'], ent['type']))
        ent_pred = set(pred['entity_list'])
        if len(ent_truth - ent_pred) != 0 or len(ent_pred - ent_truth) != 0:
            bad_cases[truth["text"]]["ent"] = (ent_truth - ent_pred, ent_pred - ent_truth)
        rel_truth = set()
        relation_truth = truth['relation_list']
        for rel in relation_truth:
            rel_truth.add((rel['subject'], rel['predicate'][1], rel['object']))
        rel_pred = set(pred['relation_list'])
        if len(rel_truth - rel_pred) != 0 or len(rel_pred - rel_truth) != 0:
            bad_cases[truth["text"]]["rel"] = (rel_truth - rel_pred, rel_pred - rel_truth)
    return bad_cases
```

```python
import re
def find_entity(data, entity_list):
    entity_dict = dict()
    text = data['data']['text']
    id_ = data['id']
    for i, (entity,typ) in enumerate(entity_list):
        if re.search(re.escape(entity), text):
            start = re.search(re.escape(entity), text).span(0)[0]
            end = re.search(re.escape(entity), text).span(0)[1] 
            text = text[:start] + '#'*len(entity) + text[end:]
            value_dict = {"start": start,
                           "end": end,
                           "text": entity,
                           "labels": [typ]}
            entity_dict[f'{id_}_{i}'] = value_dict
    return entity_dict

def find_relation(entity_dict, relation_list):
    new_relation_list = []
    entity_filter_s = []
    entity_filter_o = []   
    
    for relation in relation_list:
        s = relation[0]
        o = relation[2]
        r = relation[1]
        from_id, to_id, typ, direction, labels = None, None, 'relation', 'right', [r]
        for k,v in entity_dict.items():
            if r == 'position-include':
                if v['text'] == o and k in entity_filter_o:
                    continue
            if v['text'] == s:
                from_id = k
                entity_filter_s.append(k)
            if v['text'] == o:
                to_id = k
                entity_filter_o.append(o)
             
            if from_id and to_id:
                break
        if from_id and to_id:
            relation_dict = {
                        "from_id": from_id,
                        "to_id": to_id,
                        "type": "relation",
                        "direction": "right",
                        "labels": labels
                    }
            new_relation_list.append(relation_dict)
            
    return new_relation_list
```

```python
def add_space_before_punctuation(text):
    text1 = re.sub(r'(?<=[^\s\\>)])\((?=[^\s])', r' (', text)
    text2 = re.sub(r'(?<=[^\s])\,', r' ,', text1)
    text3 = re.sub(r'\§(?=([0-9A-Za-z]+))', r'§ ', text2)
    text4 = re.sub(r':-', r' :-', text3)
    if text4.endswith(':') and not text4.endswith(' :'):
        text4 = text4[:-1] + ' :'
    if text4.endswith('.') and not text4.endswith(' .'):
        text4 = text4[:-1] + ' .'
    text5 = re.sub(r'(?<=[^\s])\;', r' ;', text4)
    return text5

def gpt_postprocess(result, typ=1):
    if typ==1:
        cur_result = defaultdict(list)
        result_seps = result.strip(';').split(';')
        for sep in result_seps:
            try:
                sep = [sep.replace(value, key) for key, value in label2symbol.items() if value in sep][0]
            except:
                pass
    #                 print('T')
            if '<E>' in sep:  # 实体
                cur_result['entity_list'].append(tuple(sep.split('<E>')))
            else:  # 关系
                b_1 = '<S>' in sep
                b_2 = '<O>' in sep
                b_3 = '<P>' in sep
                if b_1 and b_2 and b_3:
                    sub = sep.split('<S>')[0].strip()
                    rest = sep.split('<S>')[1]
                    obj = rest.split('<O>')[0]
                    try:
                        rest = rest.split('<O>')[1]
                    except:
                        pass
                    rel = rest.replace('<P>', '').strip('.')
                    cur_result['relation_list'].append((sub, rel, obj))
    if typ==0:
        cur_result = defaultdict(list)
        entity_dict = {}
        entity_tuple_l = []
        text = add_space_before_punctuation(result)
        for k,v in entity_label.items():
            pattern = "(?<=<{}>).+?(?=</{}>)".format(str(v).replace('$', ''), str(v).replace('$', ''))
            entity_tuple_l.extend([(text[m.start(0):m.end(0)], k) for m in re.finditer(pattern,text)])
        cur_result['entity_list'] = entity_tuple_l
#         entity_tuple_l.sort(key=lambda x: x[:2])
#         for i,tuple_ in enumerate(entity_tuple_l):
#             s = tuple_[0]-3-i*7
#             e = tuple_[1]-3-i*7
#             entity_dict['{};{}'.format(str(s), str(e))] = tuple_[2]
    if typ==2:
        cur_result = defaultdict(list)
        seq = result.split('\n')
        entity_list = []
        relation_list = []
        for s in seq:
            if s.split(':')[0].split(' ')[1]== 'entity':
                label = s.split(':')[0].split(' ')[0]
                ent_l = s.split(':')[1].strip().split('; ')
                for ent in ent_l:
                    if ent:
                        entity_list.append((ent, label))
            if s.split(':')[0].split(' ')[1] == 'relation':

                label = s.split(':')[0].split(' ')[0]
                rel_l = s.split(':')[1].strip()
                if rel_l:
                    rel_l = rel_l.split('; ')
                else: 
                    rel_l = []
                for rel in rel_l:
                    if 'and <object>' in rel:
                        s = rel.split('and <object>')[0].replace('<subject>', '').strip()
                        o = rel.split('and <object>')[1].strip()
                        relation_list.append((s, label, o))
                    elif relation_list:
                        s = relation_list[-1][0]
                        o = rel.strip()
                        relation_list.append((s, label, o))
        cur_result['relation_list'] = relation_list
        cur_result['entity_list'] = entity_list
        
    return cur_result
```

```python
def calculate_metrics(pred_results, truth_results, typ=1, add_relation=True):
    metrics = defaultdict(int)
    # truth_results = remove_bad_case(truth_results)    assert len(pred_results) == len(truth_results)
    for pred, truth in zip(pred_results, truth_results):
        ent_truth = set(truth['entity_list'])
#         entity_truth = truth['entity_list']
#         for ent in entity_truth:
#             ent_truth.add((ent['text'], ent['type']))
        metrics['ent_tp'] += len(set(pred['entity_list']) & ent_truth)
        metrics['ent_fp'] += len(set(pred['entity_list']) - ent_truth)
        metrics['ent_fn'] += len(ent_truth - set(pred['entity_list']))
        rel_truth = set(truth['relation_list'])
#         relation_truth = truth['relation_list']
#         for rel in relation_truth:
#             rel_truth.add((rel['subject'], rel['predicate'][1], rel['object']))
        metrics['rel_tp'] += len(set(pred['relation_list']) & rel_truth)
        metrics['rel_fp'] += len(set(pred['relation_list']) - rel_truth)
        metrics['rel_fn'] += len(rel_truth - set(pred['relation_list']))
    ent_precision = metrics['ent_tp'] / (metrics['ent_tp'] + metrics['ent_fp'])
    ent_recall = metrics['ent_tp'] / (metrics['ent_tp'] + metrics['ent_fn'])
    ent_f1 = 2 * ent_precision * ent_recall / (ent_precision + ent_recall)
    if typ != 0 and add_relation:
        rel_precision = metrics['rel_tp'] / (metrics['rel_tp'] + metrics['rel_fp'])
        rel_recall = metrics['rel_tp'] / (metrics['rel_tp'] + metrics['rel_fn'])
        rel_f1 = 2 * rel_precision * rel_recall / (rel_precision + rel_recall)
    else:
        rel_precision, rel_recall, rel_f1 = None, None, None
    metrics['ent_precision'], metrics['ent_recall'], metrics['ent_f1'] = ent_precision, ent_recall, ent_f1    
    metrics['rel_precision'], metrics['rel_recall'], metrics['rel_f1'] = rel_precision, rel_recall, rel_f1    
    return metrics

# calculate_metrics([gpt_postprocess(data[0]['1'], typ=0)], support_dataset[0:1])
```
